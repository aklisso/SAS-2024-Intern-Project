```python
nulls_geq = (final_data5.isnull().sum(axis=0)/len(final_data5['id'])) >= 0.5
print(f'The columns with >= 50% null in the Family History data set are {final_data5.columns[nulls_geq]}')
```

```python
#since none of these appear crucial to our model we may drop them

final_data6 = final_data5.drop(columns=final_data5.columns[nulls_geq])
final_data6.head(2)
```

```python
#checking which ones have > 20% null because those will be more difficult to impute
nulls_geq20 = (final_data6.isnull().sum(axis=0)/len(final_data6['id'])) >= 0.2
print(f'The columns with >= 20% null in the Family History data set are {final_data6.columns[nulls_geq20]}')

#again, these columns appear redundant (percent cdc13 positive at different visits) so we may drop
#we will maintain pct_cd63pos_0_v60. even though it's >20% we need it because in literature it is a good predictor
cols_to_drop= list(set(final_data6.columns[nulls_geq20])-{'pct_cd63_pos_aIge'})
final_data7 = final_data6.drop(columns = cols_to_drop)
final_data7.head(2)
```

```python
final_data8 = final_data7.drop(columns = ['challenge_type', 'challenge_rxn', 'study','trt', 'leap_sev_rxn'] )
#dropping challenge_type because it wouldn't make sense as a predictor of allergy
#dropping challenge_rxn because it is the same as whether OFC indicated allergy
#dropping study because it doesn't matter
#dropping trt because they're all untreated
#dropping leap_sev_rxn because redundant

final_data8.head(2)
```

```python
#retrieving names of columns we must impute
nulls_geq0 = (final_data8.isnull().sum(axis=0)/len(final_data6['id'])) > 0
print(f'The columns with > 0% null in the Family History data set are {final_data8.columns[nulls_geq0]}')
```

```python
cat_vars = ['ofc']

cont_vars = ['pnt_IgG4_log', 'pnt_IgG4_IgE_ratio', 'pnt_spec_IgG4',
        'pnt_ara_h1', 'pnt_ara_h2', 'pnt_ara_h3',
       'pnt_ara_h8', 'pnt_ara_h9', 'pnt_ara_h1_log', 'pnt_ara_h2_log',
       'pnt_ara_h3_log', 'pnt_ara_h8_log', 'pnt_ara_h9_log', 'cum_dose_tol',
       'tot_ige']
#since pct_cd63_pos_aIge has such a high proportion of missing values, I will impute it with MICE soon.

from sklearn.impute import SimpleImputer

fill_cat = SimpleImputer(missing_values = pd.NA, strategy = "most_frequent")
fill_cont = SimpleImputer(missing_values = pd.NA, strategy = "mean")

final_data9 = final_data8

final_data9[cat_vars] = fill_cat.fit_transform(final_data9[cat_vars])
final_data9[cont_vars] = fill_cont.fit_transform(final_data9[cont_vars])
```

```python
#impoting the iterative imputer (we also need to enable it since it is in an experimental version)
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
imp = IterativeImputer(estimator = lr, tol=1e-10, max_iter = 30, verbose = 2, imputation_order='roman')

final_data9['pct_cd63_pos_aIge'] = final_data8['pct_cd63_pos_aIge'].astype(float)

final_data9['pct_cd63_pos_aIge'] = imp.fit_transform(final_data9[['pct_cd63_pos_aIge']])

```

```python
#checking all our nulls are gone
nulls = ((final_data9.isnull().sum(axis=0)/len(final_data9['id'])) > 0) 
print(f'The {len(final_data9.columns[nulls])} columns with any nulls in the Family History data set are {final_data9.columns[nulls]}')
```

```python
final_data10 = final_data9.merge(compiled_3, on = 'id', how = 'inner')
final_data10.head(2)
```

```python
#final_data10.hist(column='pnt_wheal_mm')
#Peanut wheal is still very skewed so let's log transform that too


#creating a column to log transform but first we need to address the 0's
zero_index = final_data10['pnt_wheal_mm']==0

#creating a new dataset to work with in case we make a mistake!
final_data11 = final_data10
#creating a placeholder column that will later become our log-transformed values
final_data11['pnt_wheal_log'] = final_data11['pnt_wheal_mm']

#first I am adding 1 to each value so we don't get -inf as a result of our log for 0 values.
    #this convention is recommended across various internet sources
final_data11['pnt_wheal_log'] = final_data11['pnt_wheal_log'] + 1

#now I'll log transform the column
final_data11['pnt_wheal_log'] = np.log10(final_data11['pnt_wheal_log'])

#checking to see if skewness has decreased
final_data11.hist(column='pnt_wheal_mm')
final_data11.hist(column='pnt_wheal_log')

```

```python
#might try a square root transformation instead:

final_data12 = final_data11
final_data12['pnt_wheal_sqrt'] = np.sqrt(final_data12['pnt_wheal_mm'])

#checking to see if skewness has decreased
final_data12.hist(column='pnt_wheal_mm')
final_data12.hist(column='pnt_wheal_sqrt')

#the square root looks better so i will just use that
final_data13 = final_data12.drop(columns=['pnt_wheal_log'])
final_data13.head(2)
```
